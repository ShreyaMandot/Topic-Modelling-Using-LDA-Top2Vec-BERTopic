```python
from top2vec import Top2Vec
```

```python
import pandas as pd
df_top2Vec = pd.read_csv('Clean_Tweets.csv')
top2vec_df = df_top2Vec.loc[:,['new_text']]
#tweets_df.info()

top2vec_df['new_text']= df_top2Vec['new_text'].str.lower().astype('str')

#convert to list 
doc1 = top2vec_df.new_text.to_list()
```
